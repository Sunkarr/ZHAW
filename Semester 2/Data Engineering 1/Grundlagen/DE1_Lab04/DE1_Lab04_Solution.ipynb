##### k) Which canton has the highest revenue?

```python
print("Solve on your own and we discuss the solution in the Lab")
revenue = transactions.groupby(['customer']).price.sum().sort_index(ascending=True)

df_merge = customers.merge(revenue, left_on='id', right_on='customer') 
df_merge.groupby(['canton']).price.sum().sort_values(ascending=False)
```
