```python
import random

# Generate 1000 unique random numbers between 1 and 10000
random_numbers = random.sample(range(1, 100001), 10000)

def sort_list(lst):
    sorted_list = []
    while lst:
        min_value = min(lst)
        sorted_list.append(min_value)
        lst.remove(min_value)
    return sorted_list

sorted_numbers = sort_list(random_numbers)
print(sorted_numbers)
```
