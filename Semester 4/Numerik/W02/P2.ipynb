# Aufgabe 4

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plotten Sie f(x) mit 400 ¨aquidistanten x-Werten im Intervall [−0.5 · 10 −7, 0.5 · 10 −7].

def f (x):
    return 1 - np.cos(x)

x = (np.linspace(-0.5*10**-7, 0.5*10**-7, 400))
plt.plot(x, f(x))
plt.show()
```

```python
def p(x):
    return x**6 - 12*x**5 + 60*x**4 - 160*x**3 + 240*x**2 - 192*x + 64
```

```python
x = np.linspace(1.99, 2.01, 400)
plt.plot(x, p(x))
plt.show()
```
