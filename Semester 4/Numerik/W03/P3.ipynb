# Aufgabe 2

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
f_x = lambda x: x * (1 + x)**0.5
t_1 = lambda x: np.sqrt(2) + (np.sqrt(2) + 1 / (2 * np.sqrt(2))) * (x - 1)  # Linearisierung an x_0 = 1
x_0 = 1

x = np.linspace(0, 5, 50)
plt.plot(x, f_x(x))
plt.plot(x, t_1(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()

```

# Aufgabe 4

```python
f_x = lambda x: np.exp(x)
def taylor(x, n):
    t = x * 0.0
    for i in range(n + 1):  # n+1, damit es auch wirklich n-ter Grad ist
        t += x**i / math.factorial(i)
    return t

x = np.linspace(-1, 1, 10)
plt.plot(x, f_x(x))
plt.plot(x, taylor(x, 2))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
# Fehlerabschätzung für Taylorpolynom
n_values = [5, 10, 20, 40]
x = 1
f_x_value = f_x(x)
# Fehlerabschätzung
for n in n_values:
    t = taylor(x, n)
    error = abs(f_x_value - t)
    print(f'Fehler für n = {n}: {error:.10f}')
```
