## Aufgabe 3

Aufgabe 3 (Bisektion, Newton-Verfahren)
Bestimmen Sie den Schnittpunkt von g(x) = exp(−x − 1) und h(x) = ln(x^2 + 1)
a) mit dem Newton-Verfahren mit Startwert x0 = 0
b) mit der Bisektion mit Startintervall [0, 1]
auf tol = 10−10 genau. Gegeben Sie jeweils auch die Anzahl der benötigten Iterationen an.

```python
def f_x(x):
    y = np.exp(-x-1) - np.log(x**2 + 1)
    return y

def f_x_derivative(x):
    y = -np.exp(-x-1) - 2*x/(x**2 + 1)
    return y
```

```python
x = 0
tol = 10e-10
k = 0
res = abs(f_x(x))
while res > tol:
    x = x - f_x(x)/f_x_derivative(x)
    res = abs(f_x(x))
    k=k+1
    print( x, res, k )
```

```python
a = 0
b = 1
tol = 10e-10
k = 0
res = abs(f_x(a))
while res > tol:
    c = 0.5*(a+b)
    if f_x(a)*f_x(c) < 0:
        b = c
    else:
        a = c
    res = abs(f_x(c))
    k=k+1
    print( a, b, k )
```

## Aufgabe 4 (Bisektion, Newton-Verfahren)

```python
def f_x(x):
    y = x**3 - 1.9*x**2 - 1.5*x + 5.3
    return y

def f_x_derivative(x):
    y = 3*x**2 - 3.8*x - 1.5
    return y
```

```python
x = 100
tol = 10e-8
k = 0
res = abs(f_x(x))
while res > tol:
    x = x - f_x(x)/f_x_derivative(x)
    res = abs(f_x(x))
    k=k+1
    print( x, res, k )
```
