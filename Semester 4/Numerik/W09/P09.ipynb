## Aufgabe 2

**Bestimmen Sie die Ausgleichsgeraden zu den folgenden Daten:**

### Datensatz 1:
| x   | -1 | 0   | 1   | 2   | 3   |
|-----|----|-----|-----|-----|-----|
| y   | 1  | 3.5 | 6   | 8.5 | 10  |

### Datensatz 2:
| x | -2  | 0   | 1   | 3   | 4   |
|---|-----|-----|-----|-----|-----|
| y | 5   | 6   | 8   | 8.5 | 9.5 |

**Aufgaben:**
- Stellen Sie jeweils die Daten und die Ausgleichsgerade graphisch dar.
- Geben Sie die Fehlerquadratsummen an.


```python
# Datensatz 1
x1 = np.array([-1, 0, 1, 2, 3])
y1 = np.array([1, 3.5, 6, 8.5, 10])

# Berechnung der Koeffizienten der Ausgleichsgeraden
A1 = np.vstack([np.ones(len(x1)), x1]).T
b1 = y1.reshape(-1, 1)
p1 = np.linalg.solve(A1.T @ A1, A1.T @ b1)
a1, b1 = p1[0][0], p1[1][0]
# Fehlerquadratsumme
y1_pred = A1 @ p1
error1 = np.sum((y1 - y1_pred.flatten())**2)
print(f"Datensatz 1: a = {a1}, b = {b1}, Fehlerquadratsumme = {error1}")
# Plotten der Ausgleichsgerade und der Datenpunkte
plt.figure()
plt.plot(x1, y1, 'ro', label='Datenpunkte')
plt.plot(x1, y1_pred, 'b-', label='Ausgleichsgerade')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Datensatz 1 - Ausgleichsgerade')
plt.legend()
plt.grid()
plt.show()
```

```python
# Datensatz 2
x2 = np.array([-2, 0, 1, 3, 4])
y2 = np.array([5, 6, 8, 8.5, 9.5])
# Berechnung der Koeffizienten der Ausgleichsgeraden
A2 = np.vstack([np.ones(len(x2)), x2]).T
b2 = y2.reshape(-1, 1)
p2 = np.linalg.solve(A2.T @ A2, A2.T @ b2)
a2, b2 = p2[0][0], p2[1][0]
# Fehlerquadratsumme
y2_pred = A2 @ p2
error2 = np.sum((y2 - y2_pred.flatten())**2)
print(f"Datensatz 2: a = {a2}, b = {b2}, Fehlerquadratsumme = {error2}")
# Plotten der Ausgleichsgerade und der Datenpunkte
plt.figure()
plt.plot(x2, y2, 'ro', label='Datenpunkte')
plt.plot(x2, y2_pred, 'b-', label='Ausgleichsgerade')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Datensatz 2 - Ausgleichsgerade')
plt.legend()
plt.grid()
plt.show()
```

## Aufgabe 3

**Bestimmen Sie das Ausgleichspolynom 2. Grades**
\[ p(x) = a + bx + cx^2 \]
**zu den folgenden Daten:**

| x   | -1 | 0 | 1  | 2  | 3  |
|-----|----|---|----|----|----|
| y   | 3  | 2 | 9  | 21 | 49 |

**Aufgaben:**
- Geben Sie auch die Normalengleichung an.
- Plotten Sie die Daten und das Ausgleichspolynom.

```python
# Daten
x3 = np.array([-1, 0, 1, 2, 3])
y3 = np.array([3, 2, 9, 21, 49])
# Normalengleichung
A3 = np.vstack([np.ones(len(x3)), x3, x3**2]).T
b3 = y3.reshape(-1, 1)
# Berechnung der Koeffizienten des Ausgleichspolynoms
p3 = np.linalg.solve(A3.T @ A3, A3.T @ b3)
a3, b3, c3 = p3[0][0], p3[1][0], p3[2][0]
# Fehlerquadratsumme
y3_pred = A3 @ p3
error3 = np.sum((y3 - y3_pred.flatten())**2)
print(f"Ausgleichspolynom 2. Grades: a = {a3}, b = {b3}, c = {c3}, Fehlerquadratsumme = {error3}")
# Plotten der Ausgleichspolynoms und der Datenpunkte
plt.figure()
plt.plot(x3, y3, 'ro', label='Datenpunkte')
plt.plot(x3, y3_pred, 'b-', label='Ausgleichspolynom 2. Grades')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ausgleichspolynom 2. Grades')
plt.legend()
plt.grid()
plt.show()
```

## Aufgabe 4

Fitten Sie die Funktion
\[ f(x) = a\sqrt{x + 1} + bx \]
im Sinne der kleinsten Fehlerquadrate an die folgenden Daten:

| x   | -1 | 0 | 3  | 8  | 15 |
|-----|----|---|----|----|----|
| y   | -1 | 3 | 10 | 27 | 42 |

**Aufgaben:**
- Bestimmen Sie eine Näherung für \( y \) an der Stelle \( x = 6 \).
- Geben Sie auch die Normalengleichung an.

```python
# Daten
x4 = np.array([-1, 0, 3, 8, 15])
y4 = np.array([-1, 3, 10, 27, 42])
# Normalengleichung
A4 = np.vstack([np.sqrt(x4 + 1), x4]).T
b4 = y4.reshape(-1, 1)
# Berechnung der Koeffizienten
p4 = np.linalg.solve(A4.T @ A4, A4.T @ b4)
a4, b4 = p4[0][0], p4[1][0]
# Fehlerquadratsumme
y4_pred = A4 @ p4
error4 = np.sum((y4 - y4_pred.flatten())**2)
print(f"Funktion: a = {a4}, b = {b4}, Fehlerquadratsumme = {error4}")
# Näherung für y an der Stelle x = 6
x6 = 6
y6 = a4 * np.sqrt(x6 + 1) + b4 * x6
print(f"Näherung für y an der Stelle x = {x6}: y = {y6}")
# Plotten der Funktion und der Datenpunkte
plt.figure()
plt.plot(x4, y4, 'ro', label='Datenpunkte')
plt.plot(x4, y4_pred, 'b-', label='Funktion')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funktion und Datenpunkte')
plt.legend()
plt.show()
```

## Aufgabe 6

Gegeben sind die folgenden Daten:

| xᵢ | 0   | 1   | 2   | 3   | 4   | 5   |
|----|-----|-----|-----|-----|-----|-----|
| yᵢ | 9.1 | 4.1 | 2.4 | 1.7 | 1.1 | 0.7 |

**a)** Fitten Sie die Daten nach der Methode der kleinsten Quadrate unter Verwendung des Ansatzes
\( y(x) = a \cdot \exp(-x) + b \).

**b)** Verwenden Sie nun die Ansatzfunktion
\( y(x) = \frac{c}{1 + x} + d \)
und bestimmen Sie die Parameter \( c \) und \( d \) nach der Methode der kleinsten Quadrate.

**c)** Stellen Sie die Daten zusammen mit den beiden Ausgleichsfunktionen auf dem Intervall \([0, 5]\) mit \(\Delta x = 0.1\) graphisch dar. Kennzeichnen Sie die Daten und Graphen mit Hilfe einer Legende.

**d)** Welche der beiden Funktionen gleicht die Daten besser an? Begründen Sie mit Hilfe der Fehlerquadratsumme.


```python
# Daten
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([9.1, 4.1, 2.4, 1.7, 1.1, 0.7])
# Normalengleichung für den ersten Ansatz
A1 = np.vstack([np.exp(-x), np.ones(len(x))]).T
b1 = y.reshape(-1, 1)
# Berechnung der Koeffizienten
p1 = np.linalg.solve(A1.T @ A1, A1.T @ b1)
a1, b1 = p1[0][0], p1[1][0]
# Fehlerquadratsumme
y_pred1 = A1 @ p1
error1 = np.sum((y - y_pred1.flatten())**2)
print(f"Erster Ansatz: a = {a1}, b = {b1}, Fehlerquadratsumme = {error1}")
# Normalengleichung für den zweiten Ansatz
A2 = np.vstack([1 / (1 + x), np.ones(len(x))]).T
b2 = y.reshape(-1, 1)
# Berechnung der Koeffizienten
p2 = np.linalg.solve(A2.T @ A2, A2.T @ b2)
c2, d2 = p2[0][0], p2[1][0]
# Fehlerquadratsumme
y_pred2 = A2 @ p2
error2 = np.sum((y - y_pred2.flatten())**2)
print(f"Zweiter Ansatz: c = {c2}, d = {d2}, Fehlerquadratsumme = {error2}")
# Plotten der Daten und der Ausgleichsfunktionen
x_plot = np.arange(0, 5.1, 0.1)
y_plot1 = a1 * np.exp(-x_plot) + b1
y_plot2 = c2 / (1 + x_plot) + d2
plt.figure()
plt.plot(x, y, 'ro', label='Datenpunkte')
plt.plot(x_plot, y_plot1, 'b-', label='Erster Ansatz')
plt.plot(x_plot, y_plot2, 'g-', label='Zweiter Ansatz')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ausgleichsfunktionen')
plt.legend()
plt.grid()
plt.show()
```

## Aufgabe 7

Die Temperaturabhängigkeit des elektrischen Widerstandes \( R \) eines Wolframdrahtes kann mit Hilfe der Gleichung

\[
R = R_0 \cdot [1 + a \cdot (T - T_0)]
\]

beschrieben werden. Dabei sind \( R_0 \) der Widerstand zur Temperatur \( T_0 = 20^\circ C \) und \( a \) der Temperaturkoeffizient.

| \( T \) (°C) | 20  | 100 | 180 | 260 | 340 | 420 | 500  |
|--------------|------|-----|-----|-----|-----|-----|------|
| \( R \) (Ω)  | 500  | 676 | 870 | 1060| 1205| 1410| 1565 |

### Aufgaben:

1. **Bestimmen Sie \( R_0 \) und \( a \), so dass die Daten bestmöglich im Sinne der kleinsten Fehlerquadratsumme approximiert werden.**
   Lösen Sie dazu das entsprechende Normalengleichungssystem.

2. **Bestimmen Sie \( R_0 \) und \( a \) mit Hilfe des Befehls `numpy.polyfit`.**

3. **Stellen Sie Daten und Ausgleichsfunktion graphisch dar und bestimmen Sie den Widerstand \( R \) bei \( T = 300^\circ C \):**
   - Ohne und
   - Mit Hilfe des Befehls `numpy.polyval`.

```python
# Daten
T = np.array([20, 100, 180, 260, 340, 420, 500])
R = np.array([500, 676, 870, 1060, 1205, 1410, 1565])
# Normalengleichung
A = np.vstack([np.ones(len(T)), T - 20]).T
b = R.reshape(-1, 1)
# Berechnung der Koeffizienten
p = np.linalg.solve(A.T @ A, A.T @ b)
R0, a = p[0][0], p[1][0]
# Fehlerquadratsumme
y_pred = A @ p
error = np.sum((R - y_pred.flatten())**2)
print(f"Normalengleichung: R0 = {R0}, a = {a}, Fehlerquadratsumme = {error}")
# Bestimmung von R0 und a mit numpy.polyfit
p_polyfit = np.polyfit(T - 20, R, 1)
R0_polyfit, a_polyfit = p_polyfit[1], p_polyfit[0]
print(f"numpy.polyfit: R0 = {R0_polyfit}, a = {a_polyfit}")
# Bestimmung von R bei T = 300°C
T_300 = 300
R_300 = R0 + a * (T_300 - 20)
print(f"Widerstand bei T = 300°C: R = {R_300}")
# Bestimmung von R bei T = 300°C mit numpy.polyval
R_300_polyval = np.polyval(p_polyfit, T_300 - 20)
print(f"Widerstand bei T = 300°C mit numpy.polyval: R = {R_300_polyval}")
# Plotten der Daten und der Ausgleichsfunktion
T_plot = np.arange(20, 500, 1)
R_plot = R0 + a * (T_plot - 20)
plt.figure()
plt.plot(T, R, 'ro', label='Datenpunkte')
plt.plot(T_plot, R_plot, 'b-', label='Ausgleichsfunktion')
plt.xlabel('Temperatur (°C)')
plt.ylabel('Widerstand (Ω)')
plt.title('Temperaturabhängigkeit des Widerstandes')
plt.legend()
plt.grid()
plt.show()
```
