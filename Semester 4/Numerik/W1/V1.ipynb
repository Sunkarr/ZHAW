```python
import numpy as np
import matplotlib.pyplot as plt

def plot_circle(a, b, num_points=100):
    # Stutzstellen für t im Intervall [a, b]
    t_values = np.linspace(a, b, num_points)

    # Funktionen x = cos(t) und y = sin(t)
    x = np.cos(t_values)
    y = np.sin(t_values)

    # Plot
    plt.figure(figsize=(6,6))
    plt.plot(x, y, label='Einheitskreis')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Kreisbahn: x = cos(t), y = sin(t), t ∈ [{a}, {b}]')
    plt.legend()
    plt.grid()
    plt.axis('equal')  # Gleiche Skalierung für x- und y-Achse
    plt.show()

def plot_lissajous(A, a, B, b, num_points=1000):
    # Stutzstellen für t im Intervall [0, 2π]
    t_values = np.linspace(0, 2 * np.pi, num_points)

    # Lissajous-Figuren
    x = np.sin(A * t_values + a)
    y = np.sin(B * t_values + b)

    # Plot
    plt.figure(figsize=(6,6))
    plt.plot(x, y, label=f'Lissajous-Figur: A={A}, a={a}, B={B}, b={b}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Lissajous-Figur')
    plt.legend()
    plt.grid()
    plt.axis('equal')  # Gleiche Skalierung für x- und y-Achse
    plt.show()

# Beispielaufrufe
plot_circle(0, 2 * np.pi)
plot_lissajous(2, np.pi/4, 1, 0)
plot_lissajous(3, np.pi/2, 1, 0)
plot_lissajous(8, 4, 3, 3)
```
