# Numerisches LÃ¶sen von Differentialgleichungen

```python
# Plott die Funktion f(t) = = 3 * t - t^3 / 4
import numpy as np
import matplotlib.pyplot as plt
f_t = lambda t: 3 * t - t**3 / 4
t = np.linspace(0, 6, 100)
plt.plot(t, f_t(t), label='f(t) = 3 * t - t^3 / 4')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Plot of f(t)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
