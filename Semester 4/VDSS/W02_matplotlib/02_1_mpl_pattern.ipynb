# The Matplotlib Pattern
[https://github.com/matplotlib/matplotlib](https://github.com/matplotlib/matplotlib)

```python
import matplotlib.pyplot as plt
```

```python
# makes the inline-figures in notebooks look crisper
%config InlineBackend.figure_format = 'svg'
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], [2,3,1,4,5])
plt.show()
```

## A Scatterplot

```python
fig, ax = plt.subplots()
ax.scatter(x=[1,2,3,4,5], y=[2,3,1,4,5])
plt.show()
```

## The Pyplot Interface

```python
plt.plot([1,2,3,4,5], [2,3,1,4,5])
plt.show()
```

## The Axes Interface is Prefered

- Better control
- More customisation
- Better compatibility with other packages

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], [2,3,1,4,5])
plt.show()
```
