# Customising Graphs

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

# makes the inline-figures in notebooks look crisper
%config InlineBackend.figure_format = 'svg'
```

## General Customisation Options

Adjust `color`, `edgecolor`, `alpha` and `linewidth` in the example below:

```python
color='blue'
edgecolor='none'
alpha=1.0
linewidth=2

fig = plt.figure(layout="tight")

ax1 = fig.add_subplot(2, 2, (3, 4))

ax1.plot(
  [1,2,3], [1,2,5], color=color, alpha=alpha, linewidth=linewidth
)

ax2 = fig.add_subplot(2, 2, 2)
ax2.bar(
  [1,2,3], [1,2,3], color=color, edgecolor=edgecolor,
alpha=alpha, linewidth=linewidth
)

ax3 = fig.add_subplot(2, 2, 1)
ax3.scatter(
  [1,2,3], [1,2,3], s=100, color=color,
  edgecolor=edgecolor, alpha=alpha, linewidth=linewidth
)

plt.show()
```

## Chart Type-Specific Customisation Options

The `bin` parameter in histogramms: Modify the number of bins in the example below

```python
bins = 15

# Generate a normal distribution dataset
data = np.random.normal(loc=0, scale=1, size=1000)

fig, ax = plt.subplots(figsize=(8, 6))

ax.hist(data,
  bins=bins,
  color='skyblue',
  edgecolor='black')

plt.show()
```

The `s` parameter in scatterplots to control the size of the points:

```python
df = pd.read_csv('https://drive.switch.ch/index.php/s/peFEAubn04Cm2QO/download')

df = df[df["biocapacity"] < 20]

size=10

fig, ax = plt.subplots()
ax.scatter(
  df["footprint"],
  df["biocapacity"],
  color="#69b3a2",
  edgecolor="black",
  linewidth=.5,
  alpha=.7,
  s=size

)

plt.show()
```

Bubble chart:

```python
df = pd.read_csv('https://drive.switch.ch/index.php/s/peFEAubn04Cm2QO/download')

df = df[df["biocapacity"] < 20]

size=df['biocapacity'] * 300 - 200

fig, ax = plt.subplots()
ax.scatter(
  df["footprint"],
  df["biocapacity"],
  color="#69b3a2",
  edgecolor="black",
  linewidth=.5,
  alpha=.7,
  s=size

)

plt.show()
```

Bar Chart and the `width` parameter:

```python
width=0.5

df = pd.read_csv('https://drive.switch.ch/index.php/s/OzSRiCPlF09fkqG/download')
df = df.groupby("year", as_index=False)["n"].sum()

fig, ax = plt.subplots()
ax.bar(
  df["year"], df["n"],
  color="red",
  alpha=0.5,
  linewidth=2,
  edgecolor="black",
  width=width
)
plt.show()
```
