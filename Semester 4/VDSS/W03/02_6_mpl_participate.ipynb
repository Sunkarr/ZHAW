```python
import numpy as np
```

```python
# Our data...
x = np.linspace(0, 10, 100)
y1, y2, y3 = np.cos(x), np.cos(x + 1), np.cos(x + 2)
names = ['Signal 1', 'Signal 2', 'Signal 3']
```

```python
# thre cosine signals on top of each other
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 6))
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
plt.show()
```

`To do`: Create a horizontal bar plot using `ax.barh()` on the provided data with the following specifications:
* The bars are "orange".
* The bar height is smaller than the default value.

```python
x = [10, 20, 30, 40]
labels = ["France", "Japan", "Italy", "Spain"]
```

```python
# horizontal bar plot with barh(), bars are orange, bar height < default
fig, ax = plt.subplots()
ax.barh(labels, x, color="orange", height=0.5)
plt.show()
```

`To do`: Create a scatter plot using `ax.scatter()` with the following properties:
* the points are "black"
* the point size is 100
* the point edges are "red"
* the edge width is thick

```python
x = [10, 20, 30, 40]
y = [30, 20, 80, 10]
```

```python
# scatter plot with scatter(), points are black, size=100, edges are red, thick edge width
fig, ax = plt.subplots()
ax.scatter(x, y, color="black", s=100, edgecolor="red", linewidth=2)
plt.show()
```

`To do`: Create a histogram using `ax.hist()` with the following features:

* the bar borders are "black"

```python
x = [
   10, 12, 11, 13,
   15, 17, 12, 16,
   15, 21, 9, 14,
   14, 15, 13
]
```

```python
# histogram with hist(), bars are pink, 5 bins, bar borders are black
fig, ax = plt.subplots()
ax.hist(x, bins=5, color="pink", edgecolor="black")
plt.show()
```
