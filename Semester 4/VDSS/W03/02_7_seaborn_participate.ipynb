# Seaborn Practice

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# makes the inline-figures in notebooks look crisper
%config InlineBackend.figure_format = 'svg'
```

`To do`: Use `sns.scatterplot()` to create a scatter plot with the footprint dataset.
* use the "lifexp" column for the x-axis
* use the "gdpCapita" column for the y-axis
* use the "region" column to define the colors

```python
df_fp = pd.read_csv('data/footprint.csv')
df_fp
```

```python
# create a sns.scatterplot, lifexp on x-axis, gdpCapita on y-axis, color by region
sns.scatterplot(data=df_fp, x='lifexp', y='gdpCapita', hue='region')
plt.show()
```

`To do`: Use `sns.regplot()` to create a scatter plot with a trend line using the footprint dataset.
* use the "lifexp" column for the x-axis
* use the "gdpCapita" column for the y-axis

```python
# sns.regplot, lifexp on x-axis, gdpCapita on y-axis
sns.regplot(data=df_fp, x='lifexp', y='gdpCapita')
plt.show()
```

`To do`: Utilize the `sns.jointplot()` method to combine a scatter plot with two density charts on the Iris dataset
* assign the "sepal_length" column to the x-axis
* assign the "sepal_width" column to the y-axis
* use the "species" column solely for color differentiation

```python
df_iris = sns.load_dataset("iris")
df_iris
```

```python
# sns.jointplot, sepal_length on x-axis, sepal_width on y-axis, color by species
sns.jointplot(data=df_iris, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

`To do`: Utilize the `sns.pairplot()` function to generate a correlogram on the Iris dataset.
* set `data`=df_iris
* specify a sensible attribute for `hue`

```python
# sns.pairplot, data=df_iris, hue='species'
sns.pairplot(data=df_iris, hue='species')
plt.show()
```
