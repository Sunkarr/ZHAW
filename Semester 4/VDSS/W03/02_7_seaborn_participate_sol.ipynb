`To do`: Use `sns.scatterplot()` to create a scatter plot with the footprint dataset.
* use the "lifexp" column for the x-axis
* use the "gdpCapita" column for the y-axis
* use the "region" column to define the colors

```python
df_fp = pd.read_csv('data/footprint.csv')
df_fp
```

```python
x = df_fp["lifexp"]
y = df_fp["gdpCapita"]
hue = df_fp["region"]

fig, ax = plt.subplots()
sns.scatterplot(x=x, y=y, hue=hue, ax=ax)
plt.show()
```

`To do`: Utilize the `sns.jointplot()` method to combine a scatter plot with two density charts on the Iris dataset
* assign the "sepal_length" column to the x-axis
* assign the "sepal_width" column to the y-axis
* use the "species" column solely for color differentiation

```python
df_iris = sns.load_dataset("iris")
df_iris
```

```python
x = df_iris["sepal_length"]
y = df_iris["sepal_width"]
hue = df_iris["species"]

# fig, ax = plt.subplots()
joint_grid = sns.jointplot(x=x, y=y, hue=hue)

# Access the underlying Axes objects
ax_joint = joint_grid.ax_joint
ax_marg_x = joint_grid.ax_marg_x
ax_marg_y = joint_grid.ax_marg_y

plt.show()
```

`To do`: Utilize the `sns.pairplot()` function to generate a correlogram on the Iris dataset.
* set `data`=df_iris
* specify a sensible attribute for `hue`

```python
pair_grid = sns.pairplot(data=df_iris, hue='species')

# Access the underlying Axes objects
axes = pair_grid.axes

# Example: Customize the title of the first subplot
axes[0, 0].set_title("Correlogram on the Iris Dataset")

plt.show()
```
