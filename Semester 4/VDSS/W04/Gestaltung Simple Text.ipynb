```python
import matplotlib.pyplot as plt

# create new figure
fig, ax = plt.subplots(figsize=(7.45, 4.9))  # width, height in inches

# remove chart border
for spine in ['top', 'right', 'left', 'bottom']:
        ax.spines[spine].set_visible(False)

# remove ticks
ax.tick_params(bottom=False, left=False, labelleft=False, labelbottom=False)

# remove grid background
ax.set_axis_off()

# title the plot
ax.text(0.02, 1.55,
        'Traditional Stay-at-Home Mothers\nare Becoming Less Common',
        fontsize=24,
        color='black',
        bbox={'facecolor': 'lightgrey', 'pad': 10, 'edgecolor': 'none'},
        transform=ax.transAxes)

# add explanation
ax.text(0.02, 1.4,
        'Based on children younger than 18.\n'
        'Their mothers are categorised based on employment status.',
        fontsize=12,
        color='black',
        transform=ax.transAxes)

# add key statistic for 1970
ax.text(0.02, 1.00,
        '$\\bf{41\%}$',
        fontsize=100,
        color='grey',
        transform=ax.transAxes)

ax.text(0.02, 0.85,
        'of children had a traditional\nstay-at-home mother in 1970.',
        fontsize=18,
        color='grey',
        transform=ax.transAxes)

# add key statistic for 2012 with blue emphasis
ax.text(0.02, 0.4,
        '$\\bf{20\%}$',
        fontsize=100,
        color='steelblue',
        transform=ax.transAxes)

ax.text(0.02, 0.25,
        'of children had a traditional\nstay-at-home mother in 2012.',
        fontsize=18,
        color='steelblue',
        transform=ax.transAxes)


# add source note
ax.text(0.02, 0.05,
        'Source: Pew Research Center analysis of March Current Population Surveys\n'
        'Integrated Public Use Microdata Series (IPUMS-CPS), 1971 and 2013.',
        fontsize=10,
        color='black',
        transform=ax.transAxes)


# save the figure as a JPG file
output_filename = "Gestaltung-Simple-Text_bratsjon.jpg"
plt.savefig(output_filename, format='jpg', dpi=300, bbox_inches='tight')

# show the plot
plt.show()

# Print confirmation of file creation
print(f"Exported as {output_filename}")
```
